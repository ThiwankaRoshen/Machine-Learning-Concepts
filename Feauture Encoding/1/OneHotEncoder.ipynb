# One Hot Encoding

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/test.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

after we apply One Hot Encoder to 'sex' we get bit signal in the columns of male and female

```python
df['Sex'].unique()
```

```python
ohe = OneHotEncoder()
feature_array = ohe.fit_transform(df[['Sex']]).toarray()
```

```python
feature_labels = np.array(ohe.categories_).ravel()
```

```python
df2 = pd.DataFrame(feature_array, columns=feature_labels)
```

```python
df_final = pd.concat([df, df2], axis='columns')
df_final
```
