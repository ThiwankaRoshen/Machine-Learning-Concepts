```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fifa_data = pd.DataFrame({"Year":[2012,2014,2016], "Goals":[2,5,3]})
fifa_data=fifa_data.set_index("Year")

# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time
sns.lineplot(data=fifa_data)
```

```python
spotify_data = pd.read_csv("/content/drive/MyDrive/spotify.csv", index_col="Date")
# print(spotify_data)
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Add label for horizontal axis
plt.xlabel("Date")
```

```python
ignset = pd.read_csv("/content/drive/MyDrive/ign_scores.csv", index_col="Platform")
ignset
```

```python
plt.figure(figsize=(20,6))
# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=ignset.index, y=ignset["RPG"] )

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=ignset, annot=True)
```

```python
candy = pd.read_csv("/content/drive/MyDrive/candy.csv")
candy.head()
```

```python
sns.scatterplot(x=candy['sugarpercent'], y=candy['winpercent'])
sns.regplot(x=candy['sugarpercent'], y=candy['winpercent'])
sns.scatterplot(x=candy['sugarpercent'], y=candy['winpercent'], hue=candy['caramel'])
sns.lmplot(x="sugarpercent", y="winpercent", hue="caramel", data=candy)
```

```python
sns.swarmplot(x=candy['chocolate'],
              y=candy['winpercent'])
```

```python
# sns.histplot(candy['winpercent'])
sns.kdeplot(data=candy['winpercent'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=candy['winpercent'], y=candy['sugarpercent'],kind="kde")
```
